# Understanding Reference and Copy in Python

This notebook explains references, shallow copy, and deep copy in Python, with executable examples.

## 1. Reference Assignment

Assignment does not create a copy. Both variables point to the same object in memory.

```python
a = [1, 2, 3]
b = a
b[2] = 11

print(a)
print(b)
```

## 2. Shallow Copy

A shallow copy creates a new container, but nested objects are still shared.

```python
import copy

a = [1, 2, [3, 4]]
b = copy.copy(a)
b[2][0] = 99

print(a)
print(b)
```

## 3. Deep Copy

A deep copy recursively duplicates all nested objects.

```python
import copy

a = [1, 2, [3, 4]]
b = copy.deepcopy(a)
b[2][0] = 99

print(a)
print(b)
```
